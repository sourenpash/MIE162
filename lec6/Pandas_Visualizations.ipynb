#### Import necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

#### Import data

df.head() shows the first 5 rows of the data frame

```python
df1 = pd.DataFrame(np.random.randn(50,4), columns=['A','B','C','D'])
df1.head()
```

```python
df2 = df1 + 10
df2.head()
```

## Style

Matplotlib has several [style sheets](http://matplotlib.org/gallery.html#style_sheets) that can be used to alter the appearance of a plot. Just import matplotlib and use *plt.style.use()* prior to drawing a plot.

```python
plt.style.use('bmh')
```

## Plot Types

There are several plot types built-in to pandas:

- df.plot.area     
- df.plot.barh     
- df.plot.density  
- df.plot.hist     
- df.plot.line     
- df.plot.scatter
- df.plot.bar      
- df.plot.box      
- df.plot.hexbin   
- df.plot.kde      
- df.plot.pie

These can also be called using the kind argument with plot, e.g. for hist, *df.plot(kind='hist')*. To make other plots with this style syntax, just set kind equal to one of the key terms in the list above (e.g., 'box','barh', etc.)

## Histogram

Use *df['col_name'].plot.hist()* to plot a histogram of count values.

```python
df1['A'].plot.hist()
```

```python
df1['A'].plot.hist(rwidth=0.9) # set width of each rectangle
```

```python
df1['A'].plot.hist(edgecolor="black") # set edge color to "black"
```

```python
plt.style.use('dark_background') # change style
```

```python
df1['B'].hist()
```

```python
plt.style.use('ggplot')
```

```python
df1['A'].plot.hist(bins=50)
```

## Area

```python
df2.plot.area(alpha=0.4)
```

## Barplots

```python
df2.loc[0:10].plot.bar() # df.loc[i:j] accesses a group of rows from row i to row j
```

### Stacked Bar plots

```python
df2.loc[0:10].plot.bar(stacked=True)
```

## Line Plots

```python
df1.plot.line(y='B',figsize=(12,3),lw=1)
```

## Scatter Plots

```python
df1.plot.scatter(x='A',y='B')
```

### Color based on another column
You can color data to represent a third axis by using the 'c' argument. Here we color the points based on the 'C' column

```python
df1.plot.scatter(x='A',y='B',c='C',cmap='coolwarm')
```

### Size based on another column
You can size datapoints to represent a third axis.. Here we set the size of the points based on the values of the 'C' column.

```python
with np.errstate(divide='ignore',invalid='ignore',over='ignore',under='ignore'):
    df1.plot.scatter(x='A',y='B',s=np.abs(df1['C'])*200) # dot size is determined by the absolute value of 'C'
```

## BoxPlots

```python
df2.plot.box()
```

## Hexagonal Bin Plot

This for useful for bivariate data.

```python
df3 = pd.DataFrame(np.random.randn(2000, 2), columns=['A', 'B'])
df3.plot.hexbin(x='A',y='B',gridsize=25,cmap='Oranges')
```

## Kernel Density Estimation plot (KDE)

```python
df2['B'].plot.kde()
```

```python
df1.plot.density()
```
